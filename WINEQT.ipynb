```python

import pandas as pd
import numpy as np
import statistics as stats

data1 = pd.read_csv('WineQT.csv')

data1 = data1.drop(['fixed acidity', 'citric acid'], axis=1)

va = np.mean(data1.loc[(~data1['volatile acidity'].isna()), 'volatile acidity'])
print("va:",va)
data1.loc[(data1['volatile acidity'].isna()), 'volatile acidity'] = va

rs = np.mean(data1.loc[(~data1['residual sugar'].isna()), 'residual sugar'])
print("rs:",rs)
data1.loc[(data1['residual sugar'].isna()), 'residual sugar'] = rs

cl = np.mean(data1.loc[(~data1['chlorides'].isna()), 'chlorides'])
print("cl:",cl)
data1.loc[(data1['chlorides'].isna()), 'chlorides'] = cl

fsd = np.mean(data1.loc[(~data1['free sulfur dioxide'].isna()), 'free sulfur dioxide'])
print("fsd:",fsd)
data1.loc[(data1['free sulfur dioxide'].isna()), 'free sulfur dioxide'] = fsd

tsd = np.mean(data1.loc[(~data1['total sulfur dioxide'].isna()), 'total sulfur dioxide'])
print("tsd:",tsd)
data1.loc[(data1['total sulfur dioxide'].isna()), 'total sulfur dioxide'] = tsd

den = np.mean(data1.loc[(~data1['density'].isna()), 'density'])
data1.loc[(data1['density'].isna()), 'density'] = den
print("den:",den)


ph = np.mean(data1.loc[(~data1['pH'].isna()), 'pH'])
print("ph:",ph)
data1.loc[(data1['pH'].isna()), 'pH'] = ph

sul = np.mean(data1.loc[(~data1['sulphates'].isna()), 'sulphates'])
print("sul:",sul)
data1.loc[(data1['sulphates'].isna()), 'sulphates'] = sul

alc = np.mean(data1.loc[(~data1['alcohol'].isna()), 'alcohol'])
print("alcohol:" ,alc)
data1.loc[(data1['alcohol'].isna()), 'alcohol'] = alc

print( data1.info())
data1.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('WineQT.csv')

plt.figure(figsize=(16, 20))

plt.subplot(3, 2, 1)
plt.scatter(data['volatile acidity'], data['alcohol'], alpha=0.6, edgecolors='black', color='blue')
plt.title("Volatile Acidity vs Alcohol", fontsize=12)
plt.xlabel("Volatile Acidity")
plt.ylabel("Alcohol")


plt.subplot(3, 2, 2)
plt.plot(data['pH'], color='green', marker='o', linestyle='dashed', markersize=4)
plt.title("pH Level Trends", fontsize=12)
plt.xlabel("Sample Index")
plt.ylabel("pH Level")


plt.subplot(3, 2, 3)
plt.hist(data['residual sugar'], bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.title("Distribution of Residual Sugar", fontsize=12)
plt.xlabel("Residual Sugar")
plt.ylabel("Frequency")


plt.subplot(3, 2, 4)
plt.scatter(data['free sulfur dioxide'], data['total sulfur dioxide'], alpha=0.6, edgecolors='black', color='red')
plt.title("Free vs. Total Sulfur Dioxide", fontsize=12)
plt.xlabel("Free Sulfur Dioxide")
plt.ylabel("Total Sulfur Dioxide")

plt.subplot(3, 2, 5)
plt.hist(data['alcohol'], bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title("Alcohol Content Distribution", fontsize=12)
plt.xlabel("Alcohol")
plt.ylabel("Frequency")


plt.subplots_adjust(hspace=0.5, wspace=0.3)
plt.suptitle("Wine Quality Data Visualizations", fontsize=14, y=0.98)

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

file = pd.read_csv("WineQT.csv")

X = file.iloc[:, :-1].values  
y = file.iloc[:, -1].values   

y = np.where(y >= 7, 1, 0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

def find_optimal_k(X_train, X_test, y_train, y_test, max_k=20):
    k_values = list(range(1, max_k + 1))
    accuracies = []
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, p=1)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        accuracies.append(accuracy_score(y_test, y_pred))
    
    optimal_k = k_values[np.argmax(accuracies)]
    return optimal_k

optimal_k = find_optimal_k(X_train, X_test, y_train, y_test)
print(f"\nOptimal k value found: {optimal_k}")

model = KNeighborsClassifier(n_neighbors=optimal_k, p=1)
model.fit(X_train, y_train)

pred = model.predict(X_test)

accuracy = accuracy_score(y_test, pred)
print(f"\nTest Accuracy with k={optimal_k}: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, pred))
```
